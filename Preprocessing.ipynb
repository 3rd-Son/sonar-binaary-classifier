```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### *Load the dataset*

```python
data = pd.read_csv('sonar.csv', header = None)
data
```

```python
def plot_ts(df):
    """
    plot time history of for different variables 
    ========================================
    input: (df) Dataframe you whish to plot the time series for its columns
    """
    ts = df.copy() # declare a copy
    time = np.arange(len(ts))
    cols = ts.columns
    
    # plotting
    fig, axes = plt.subplots(len(cols), 1, figsize=(10, df.shape[1]))
    for col, ax in zip(cols, axes):
        ax.plot(ts[col],label=col)
        ax.legend(loc=2)  
    plt.tight_layout()      
    # figure title    
    # fig.suptitle('Visaulization/Observation for all variables')
    
```

```python
plot_ts(data)
```
